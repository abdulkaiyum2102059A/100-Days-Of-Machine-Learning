# Literals

Literal is a raw data given in a variable , there are various types of literals they are as follows :

- Numeric Literals
- String Literals
- Boolean Literals
- Special Literals

# 1. Numeric Literals

```python
a = 0b1000 #Binary Literals
b = 100 #Decimal Literals
c = 0o310 #Octal Literal
d = 0x12c #Hexadecimal Literal

#Float Literal
float_1 = 10.5
float_2 = 13.5e2
float_3 = 15.5e-2

#Complex Literal
x = 3.14j

print (a,b,c,d)
print(float_1,float_2,float_3)
print(x,x.imag,x.real)
```

```python
string = 'This is python'
strings = "This is python"
char = "c"
multiline_str = """This is a multipleline string with more than one line code."""
raw_str = r"raw \n string"
unicode = u"\U0001f600\U0001F606\U0001F923"

print(string)
print(strings)
print(char)
print(multiline_str)
print(unicode)
print(raw_str)
```
